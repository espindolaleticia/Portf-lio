```python
#leitura dos valores
a=float (input('Digite sua nota 1:'))
b=float (input('Digite sua nota 2:'))
c=float (input('Digite sua nota 3:'))
```

```python
media = (a*2 + b*3 + c*5) / 10
```

```python
print (f'A média do aluno é {media:.1F}\nsendo assim o aluno foi aprovado')
```
